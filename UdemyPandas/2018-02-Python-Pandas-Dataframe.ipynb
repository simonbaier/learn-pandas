```python
# rank() method
```

```python
# reset nba to original import 
nba = pd.read_csv("data/nba.csv")
nba.tail()
```

```python
nba = nba.dropna(how = "all")    # drop rows where all values are NaN
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba.head(5)
```

```python
# rank salaries
nba["Salary"].rank()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype(int)
nba.head()
```

```python
nba.sort_values(by = "Salary", ascending=False).head(20)
# note that same values share a rank number 
```
