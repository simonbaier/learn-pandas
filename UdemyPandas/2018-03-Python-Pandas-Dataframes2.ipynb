# extract duplicates and uniques 

with duplicates() method


```python
# reset 
df = pd.read_csv("data/employees.csv", parse_dates=["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
# start by sorting 
df.sort_values("First Name", inplace=True)
df.head(10)
```

```python
df["First Name"].head(10)
```

```python
# identify duplicate First Names 
df["First Name"].duplicated().head(10)  # by default duplicated(keep="first")
```

```python
# Note above how the first Aaron, Adam, Alan are not considered a duplicate, but the remaining are duplicates
# to reverse that use ...
df["First Name"].duplicated(keep="last").head(10)
# Note below how the first instance of a value is considered a duplicate, but the last is not 
```

```python
# a condensed version
df.duplicated("First Name", keep = "last").head(10)
```

```python
# now just display ALL duplicates 
df[df["First Name"].duplicated(keep = False)].head(10)
```

```python
# display unique names 
# first filter all duplicates 
df["First Name"].duplicated(keep = False)  # filters all duplicates ... but we want the NOT duplicates
# so how to do the reversal ?
# the tilde reverses booleans 
mask = ~df["First Name"].duplicated(keep = False)   # NOTE THE TILDE 
df[mask]  # displays employees with unique first names 
```

```python
# drop_duplicates() method 
# this method can be used on a data frame unlike the duplicates() method
# ... so can accomplish some of the above functions more easily
```

```python
# reset 
df = pd.read_csv("data/employees.csv", parse_dates=["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype(bool)
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace=True)
df.head(10)
```

```python
# how many rows ?
len(df)
```

```python
len(df.drop_duplicates())
```

```python
# why ?
# only works like this where rows are identical 
```

```python
# the subset parameter specifies a python list of which columns or column to consider - they are additive
df.drop_duplicates(subset=["First Name"], keep=False)  # displays just employees with unique firt names
```
