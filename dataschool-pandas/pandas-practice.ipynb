1. create a pandas dataframe named 'movies' containing IMDB movied data stored as a CCV at a bitly URL 'bit.ly/imdbratings'



```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

1. sort a series based on title

display a series of movie titles sorted in ascending order using both dot notation and bracket notation

```python
movies.title.sort_values().head(20)
# movies['title'].sort_values() # alt method
```

sort in descending order using bracket method to specify title column

```python
movies['title'].sort_values(ascending=False).head(20)
# returns a series
```

### sort a dataframe
sort a dataframe by the titles column

```python
movies.sort_values('title').head(20)
# returns full dataframe
```

### sort by multiple columns


```python
movies.sort_values(['content_rating','title'])
```
